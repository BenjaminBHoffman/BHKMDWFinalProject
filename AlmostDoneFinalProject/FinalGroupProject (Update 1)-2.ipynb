# Metric 2: Crime Rate 

The second criteria we will consider is the crime rate of each neighborhood. This factor is extremely important because as resources dwindle, people will start to go to extremes like stealing things and breaking into homes. This is why the safety of each neighborhood is important so you can ensure that you will stay safe during this virus outbreak.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt  
%matplotlib inline 

# Read the crime data from the CSV file
crime_data = pd.read_csv("updatedCrime.csv")

# Create a DataFrame to count the number of arrests by neighborhood
arrests = crime_data["INCIDENTNEIGHBORHOOD"].value_counts().head(25)

# Set the size of the bar graph
fig, ax = plt.subplots(figsize=(8, 10))

# Plot the top 25 neighborhoods with the highest number of arrests
arrests.plot(kind="barh", ax=ax, color='#e32636')
plt.xlabel("Number of Arrests")
plt.ylabel("Neighborhood")
plt.title("Top 25 Neighborhoods with the Highest Number of Arrests (August 2016 -> Current Day)")
plt.show()
```


This next bar graph shows the **25** neighborhoods with the highest crime rate. 

**Top 3 Neighborhoods With the Most Amount of Arrests**
- 1st - Central Business District
- 2nd - South Side Flats
- 3rd - Carrick

```python
# Create a DataFrame to count the number of arrests by neighborhood
arrests = crime_data["INCIDENTNEIGHBORHOOD"].value_counts().tail(25)

# Set the size of the bar graph
fig, ax = plt.subplots(figsize=(8, 10))

# Plot the top 25 neighborhoods with the lowest number of arrests
arrests.plot(kind="barh", ax=ax, color='#228b22')
plt.xlabel("Number of Arrests")
plt.ylabel("Neighborhood")
plt.title("Top 25 Neighborhoods with the Least Number of Arrests (August 2016 -> Current Day)")
plt.show()
```

This next bar graph shows the **25** neighborhoods with the lowest crime rate. 

**Top 3 Neighborhoods With the Lowest Amount of Arrests**
- 1st - Troy Hill-Herrs Island
- 2nd - Central Northside
- 3rd - Regent Square

The issue now is that this does not take into consideration the population of each neighborhood. The neighborhoods with the lower population are going to, on average, have lower crime rates and the same goes for the higher populated ones. What I did is to get another data set that has the population of all of the neighborhoods and then make it a ratio of the crime rate per 100 people. I did this by dividing the total arrests by the total population in each neighborhood and then multiply that by 100.

```python
import pandas as pd

# Read population data
neighborhoodPopulation = pd.read_csv("updatedNeighborhoodPopulation.csv")
populationDF = neighborhoodPopulation[['Neighborhood', 'Population (2010)']]

# Rename the "Population (2010)" column to "Population"
populationDF = populationDF.rename(columns={'Population (2010)': 'Population'})

# Read crime data
arrests = crime_data["INCIDENTNEIGHBORHOOD"].value_counts()
crimeDF = pd.DataFrame(arrests).reset_index()
crimeDF.columns = ['Neighborhood', 'Arrests']

# Merge dataframes
mergeDF = pd.merge(populationDF, crimeDF, on='Neighborhood')

# Print the merged dataframe
mergeDF.head()
```

Now that all of the data of the population and crimes has been put into one dataframe, we can now find the **crime rate** in each neighborhood **per 100 people**. To do this, we need to divide the number of arrests that occured by the population of the neighborhood and then multiply by 100 for it to become the crime rate per every 100 people. What I basically did was **standardize the data** where the highest and lowest crime rates wont be just mainly neighborhoods with the most and least amount of people.

```python
mergeDF["Crime Rate (Per 100 People)"] = mergeDF["Arrests"] / mergeDF["Population"] * 100

mergeDF.head()
```

Now this is a better representation of truly how dangerous each neighborhood is. One slight issue that you might see is that for a place like Allegheny Center, the total crimes is over the population and the reason for that is because the total crimes are from 2016 to current day which means that the crimes can accumulate over time and surpass the total population number.

```python
# Sort the DataFrame by CrimeRatio in descending order
mergeDF = mergeDF.sort_values(by='Crime Rate (Per 100 People)', ascending=False)

# Plot the bar graph
plt.figure(figsize=(12, 18))
plt.barh(mergeDF['Neighborhood'], mergeDF['Crime Rate (Per 100 People)'], color='#8886F5')
plt.xlabel('Crime Rate (Per 100 People)')
plt.ylabel('Neighborhood')
plt.title('Crime Rate Ratio by Neighborhood')
plt.show()
```

Holy Smokes, those numbers are crazy. It looks like we have a couple outliers that are skewing the data, making it hard to read. **Chateau**, **South Shore**, and **North Shore** all have much higher crime rates than the rest of the neighborhoods. What we can do to fix this is to remove those outliers to be able to compare the other neighborhoods better. So lets do it!!!

```python
filteredDF = mergeDF.nlargest(3, 'Crime Rate (Per 100 People)', 'all')
filteredDF = mergeDF[~mergeDF['Neighborhood'].isin(filteredDF['Neighborhood'])]

# Sort the DataFrame by Crime Ratio in descending order
filteredDF = filteredDF.sort_values(by='Crime Rate (Per 100 People)', ascending=False)
# Plot the bar graph
plt.figure(figsize=(12, 18))
plt.barh(filteredDF['Neighborhood'], filteredDF['Crime Rate (Per 100 People)'], color='#8886F5')
plt.xlabel('Crime Rate (Per 100 People)')
plt.ylabel('Neighborhood')
plt.title('Crime Rate Ratio by Neighborhood')
plt.show()
```

Now everything looks more clear. However, we dont have the exact numbers so why dont we make a chart that shows the 10 best neighborhoods with the exact crime rate per 100 showen for each one.

```python
# Sort the DataFrame by Crime Ratio in ascending order to get the lowest crime rates
filteredDF = mergeDF.sort_values(by='Crime Rate (Per 100 People)', ascending=True).head(10)

# Plot the horizontal bar chart with crime rate numbers
plt.figure(figsize=(12, 8))
bars = plt.barh(filteredDF['Neighborhood'], filteredDF['Crime Rate (Per 100 People)'], color='#02C935')

# Display crime rate numbers next to each bar
for bar, rate in zip(bars, filteredDF['Crime Rate (Per 100 People)']):
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2, f'{rate:.2f}', ha='left', va='center')

plt.xlabel('Crime Rate (Per 100 People)')
plt.ylabel('Neighborhood')
plt.title('Top 10 Neighborhoods with the Lowest Crime Rate')
plt.show()
```

 Now that looks a lot better. Here you can see all of the crime rates for the 10 best neighborhoods. Now to get to the results...

# Top 3 best neighborhoods with the lowest crime rate
- 1st - Central Northside (Crime Rate: 0.79)
- 2nd - Squirrel Hill North (Crime Rate: 2.53)
- 3rd - Swisshelm Park. (Crime Rate: 3.09)

# Metric 3: Fire Incidents

The last set of data we will consider is fire incident reports. It is crucial for surviving a dealy virus outbreak that whereever your living isn't very susceptible to fires. On first thought, it will probably be more beneficial to be in a neighborhood where the houses are less compact and for most if not all the houses to be built out of non-flamible material such as brick or stone. This data is comprised on fires incidents reported in the Pittsburgh area, which is updated daily. 

```python
import pandas as pd
fire = pd.read_csv("data/fireIncident.csv")
fire.head(10)
```

This is the full data, now lets clean it up to make it easier to interpret. 

```python
tabs = fire[['_id', 'neighborhood']]
cleaned_tabs = tabs.dropna()
cleaned_tabs.head(10)
```

Now that we have a complete and trimmed down data set to look at, lets graph this data to make it easier to visualize. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
fire = pd.read_csv("fireIncident.csv")

# Assuming 'neighborhood' is the column containing neighborhood names
# and '_id' is the column containing the frequency of fire incidents
neighborhood_data = fire.groupby('neighborhood')['_id'].sum()

# Plotting the bar graph
neighborhood_data.plot.bar(color = ['blue'], width = .75, fontsize = 13, figsize = (30,30), rot = 90)

# Adding labels and title
plt.xlabel('Neighborhoods')
plt.ylabel('Frequency of Fires')
plt.title('Frequency of Fire Incidents in Different Neighborhoods')

# Display the plot
plt.show()
```

As we can see from the graph, all of the different neighborhoods of pittsburgh have bene graphed and plotted with their respective frequencies, with the name of the neighbohood being on the x axis, and the frequency of fires being reported on the y axis. We can see from the graph that the neighborhoods with the highest rate of fire incidents reported are Carrick, Central Business District, South Side Flats, and Squirrel Hill South. The neighborhoods with the lowest frequency of fires are Mount Oliver Borough, Regent Square, and Ridgmont.
