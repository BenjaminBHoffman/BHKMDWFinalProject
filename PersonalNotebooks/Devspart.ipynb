# Lets take a look at a dataset regarding housing in Pittsburgh. 

We should keep in mind that this data is from 2010, so it may not be fully accurate to what the data today would look like.

```python
import pandas as pd
population = pd.read_csv("data/neighborhoodPopulation.csv")
population.head(20)
```

# Neighborhood dataset overview:

Even though this is only a small part of the whole data set, we can already start to get a good idea of where we might want to be and where we'd probably want to stay away from. Basically the only 2 columns that are important here are the neighborhood and population, so let's get rid of everything else. 

```python
necessary = population[['Neighborhood', 'Population (2010)']]
cleaned_data = necessary.dropna()
cleaned_data.head(30)
```

Here is a much cleaner version of the data set we were just looking at. However, this alone still doesn't help us that much because it is sorted in alphabetical order based on the names of the neighborhoods. So let's sort it by population.
