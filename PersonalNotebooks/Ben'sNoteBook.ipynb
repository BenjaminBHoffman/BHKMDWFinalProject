This data is comprised on fires incidents reported in the Pittsburgh area, which is updated daily.

Now that we have a complete and trimmed down data set to look at, lets graph this data to make it easier to visualize. 

```python
import pandas as pd
import matplotlib.pyplot as plt

fire = pd.read_csv("fireIncident.csv")

neighborhood_data = fire.groupby('neighborhood')['_id'].sum()

neighborhood_data.plot.bar(color = ['blue'], width = .75, fontsize = 13, figsize = (30,30), rot = 90)

plt.xlabel('Neighborhoods')
plt.ylabel('Frequency of Fires')
plt.title('Frequency of Fire Incidents in Different Neighborhoods')

plt.show()
```

As we can see from the graph, all of the different neighborhoods of pittsburgh have bene graphed and plotted with their respective frequencies, with the name of the neighbohood being on the x axis, and the frequency of fires being reported on the y axis. We can see from the graph that the neighborhoods with the highest rate of fire incidents reported are Carrick, Central Business District, South Side Flats, and Squirrel Hill South. The neighborhoods with the lowest frequency of fires are Mount Oliver Borough, Regent Square, and Ridgmont.
