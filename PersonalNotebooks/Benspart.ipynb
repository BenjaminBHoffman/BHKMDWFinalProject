```python
import pandas as pd
fire = pd.read_csv("data/fireincident.csv")
fire.head(10)
```

This is the full data, now lets clean it up to make it easier to interpret. 

```python
tabs = fire[['_id', 'neighborhood']]
cleaned_tabs = tabs.dropna()
cleaned_tabs.head(10)
```

Now that we have a complete and trimmed down data set to look at, lets graph this data to make it easier to visualize. 

```python
import matplotlib.pyplot as plt

neighborhood = fire.iloc[:, 0].sum()
_id_total = fire.iloc[:, 1].sum()

plt.bar(['neighborhood', '_id'], [neighborhood_total, _id_total])

plt.xlabel('Neighborhoods')
plt.ylabel('Frequency of Fires')
plt.title('Frequency of fire incidents in different neighborhoods in pittsburgh')

plt.show()
```

```python
import matplotlib.pyplot as plt

data = cleaned_tabs.sort_values(by='_id', ascending=True)
# data.set_index('Neighborhood', inplace=True)
plotted = data.plot.bar(color = ['blue'], width = .75, fontsize = 13, figsize = (30,30), rot = 90)
plt.xlabel(xlabel = 'Neighborhoods', fontsize = 25)
plt.ylabel(ylabel = 'Population (2010)', fontsize = 25)
plt.title(label = "Top 20 Pitt Neighborhoods Based on Population",fontsize = 30, pad = 25)
plt.show()
```
